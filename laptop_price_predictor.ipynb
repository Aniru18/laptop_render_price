```python
import pandas as pd

# Assuming df is your DataFrame

# Step 1: Clean the 'Memory' column
df['Memory'] = df['Memory'].astype(str).replace('\.0', '', regex=True)
df["Memory"] = df["Memory"].str.replace('GB', '')
df["Memory"] = df["Memory"].str.replace('TB', '000')

# Step 2: Split the 'Memory' column
new = df["Memory"].str.split("+", n=1, expand=True)

df["first"] = new[0].str.strip()
df["second"] = new[1].str.strip()

# Step 3: Create layers for storage types in both parts
df["HDD_1"] = df["first"].apply(lambda x: int(x.split()[0]) if "HDD" in x else 0)
df["SSD_1"] = df["first"].apply(lambda x: int(x.split()[0]) if "SSD" in x else 0)
df["Hybrid_1"] = df["first"].apply(lambda x: int(x.split()[0]) if "Hybrid" in x else 0)
df["Flash_Storage_1"] = df["first"].apply(lambda x: int(x.split()[0]) if "Flash Storage" in x else 0)

df["HDD_2"] = df["second"].apply(lambda x: int(x.split()[0]) if pd.notna(x) and "HDD" in x else 0)
df["SSD_2"] = df["second"].apply(lambda x: int(x.split()[0]) if pd.notna(x) and "SSD" in x else 0)
df["Hybrid_2"] = df["second"].apply(lambda x: int(x.split()[0]) if pd.notna(x) and "Hybrid" in x else 0)
df["Flash_Storage_2"] = df["second"].apply(lambda x: int(x.split()[0]) if pd.notna(x) and "Flash Storage" in x else 0)

# Step 4: Sum up storage types from both parts
df["HDD"] = df["HDD_1"] + df["HDD_2"]
df["SSD"] = df["SSD_1"] + df["SSD_2"]
df["Hybrid"] = df["Hybrid_1"] + df["Hybrid_2"]
df["Flash_Storage"] = df["Flash_Storage_1"] + df["Flash_Storage_2"]

# Drop the temporary 'first', 'second', and all columns with 'Layer' in their names
df.drop(columns=['first', 'second', 'HDD_1', 'HDD_2', 'SSD_1', 'SSD_2', 'Hybrid_1', 'Hybrid_2', 'Flash_Storage_1', 'Flash_Storage_2'], inplace=True)

# View the result
print(df[['Memory', 'HDD', 'SSD', 'Hybrid', 'Flash_Storage']])
```

## Applying Linear Regression

```python

from sklearn.ensemble import VotingRegressor,StackingRegressor

step1 = ColumnTransformer(transformers=[
    ('col_tnf',OneHotEncoder(sparse=False,drop='first'),[0,1,7,10,11])
],remainder='passthrough')


estimators = [
    ('rf', RandomForestRegressor(n_estimators=350,random_state=3,max_samples=0.5,max_features=0.75,max_depth=15)),
    ('gbdt',GradientBoostingRegressor(n_estimators=100,max_features=0.5)),
    ('xgb', XGBRegressor(n_estimators=25,learning_rate=0.3,max_depth=5))
]

step2 = StackingRegressor(estimators=estimators, final_estimator=Ridge(alpha=100))

pipe = Pipeline([
    ('step1',step1),
    ('step2',step2)
])

pipe.fit(X_train,y_train)

y_pred = pipe.predict(X_test)

print('R2 score',r2_score(y_test,y_pred))
print('MAE',mean_absolute_error(y_test,y_pred))
```
